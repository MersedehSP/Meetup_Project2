```python
import pandas as pd
import re
```

# Read in Scrape file from Meetup

```python
datafile_df = pd.DataFrame() 
scrapefile="complete_events_lat_lng.csv"
scrapefile_df=pd.read_csv(scrapefile)
datafile_df=scrapefile_df[["event_name","group_name","venue_event_link","attendees",
                           "Google_Map_Link","city","event_city","event_state","event_street","event_lat","event_lng"]]
datafile_df
```

# Clean up dataframe of special characters and rename columns

```python
def strip_character(dataCol):
    r = re.compile(r'[^a-zA-Z !@#$%&*_+-=|\:";<>,./()[\]{}\']')
    return r.sub('', dataCol)

pd.set_option('mode.chained_assignment', None)
datafile_df["event_name"] = datafile_df["event_name"].apply(strip_character)
datafile_df["group_name"] = datafile_df["group_name"].apply(strip_character)
datafile_df["venue_event_link"] = datafile_df["venue_event_link"].apply(strip_character)
datafile_df["Google_Map_Link"] = datafile_df["Google_Map_Link"].apply(strip_character)
datafile_df["event_street"] = datafile_df["event_street"].apply(strip_character)

datafile_df=datafile_df.rename(columns={"event_name":"evt","group_name":"grp","venue_event_link":"lnk","Google_Map_Link":"maploc","event_city":"subcity"})
datafile_df=datafile_df.rename(columns={"event_street":"street","event_lat":"subcity_lat","event_lng":"subcity_lng","event_state":"state"})
datafile_df=datafile_df.rename(columns={"attendees_num":"attendees"})
datafile_df = datafile_df.rename_axis("event_id", axis="columns")

datafile_df
```

# Read in Scrape file and create and rename columns in city dataframe 

```python
city_df = pd.DataFrame()
scrapefile="complete_events_lat_lng.csv"
scrapefile_df=pd.read_csv(scrapefile)
city_df=scrapefile_df[['city',"event_state","lat","lng","population"]]
city_df=city_df.rename(columns={"city":"city","event_state":"state","lat":"city_lat","lng":"city_lng","population":"city_population"})
city_df=city_df.drop_duplicates(ignore_index=True)
#pd.set_option("display.max_row", None, "display.max_columns", None)
city_df['city_id']=city_df.index
city_df
```

# Do a left merge to obtain the city id of subcity where the event is located

```python
events_df=pd.merge(datafile_df,city_df,how='left',on=['city','state'])
events_df['city_id'] = events_df['city_id'].astype(int)
events_df=events_df.drop(["city_lat","city_lng","city_population"], axis='columns')
events_df
```

# Save dataframes into csv to be loaded into postgresql database

```python
events_df.to_csv("events.csv",index=False)
city_df.to_csv("city.csv",index=False)
```
