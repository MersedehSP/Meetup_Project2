```python
#Dependencies
import pandas as pd
import datetime
import numpy as np
```

```python
file_path = "petrocks_finalllll4k.csv"
wine_df = pd.read_csv(file_path, encoding='ISO-8859-1')
wine_df.head(80)
```

```python
#renaming the columns to matc with everyone
wine_df.rename(columns = {'region':'state'}, inplace = True) 
wine_df.rename(columns = {'sub_region':'region'}, inplace = True)
wine_df.rename(columns = {'grape_varietal':'varietal'}, inplace = True) 
wine_df.rename(columns = {'sale_price':'price'}, inplace = True) 
```

```python
#dropping unwanted columns
#wine_df.drop(columns = {'region':'state'}, inplace = True) 
wine_df = wine_df.drop(['wine_type','bottle_price',], axis='columns')
```

```python
#droppinig the empty rows
wine_df = wine_df.dropna(how = "all")
wine_df
```

```python
len(wine_df)
```

```python
wine_df = wine_df.drop_duplicates()
wine_df
```

```python
len(wine_df)
```

```python
#some other weird duplicates needed to be removed (because I didnt use the same code for country as  did for sub region, varietal and description)
duplicated_rows = wine_df.loc[wine_df["country"]== "click for more details"]
duplicated_rows
```

```python
corrected_wine_df = wine_df[wine_df["country"]!= "click for more details"]

corrected_wine_df
len(corrected_wine_df)
```
