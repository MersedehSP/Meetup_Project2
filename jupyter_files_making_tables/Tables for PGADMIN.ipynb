```python
test1 = event_df.groupby([ 'state', 'city', 'event_city', 'lat', 'lng', 'population']).size().reset_index().rename(columns = {0:'count'})
test2 = test1.reset_index()
test2
test3 = test2.rename(columns={"index": 'region_id'})
test3 = test3.drop(["count"], axis= "columns")
test3
```

```python
new_df = pd.merge(event_df, test3, how='left', left_on=[ 'state', 'city', 'event_city', 'lat', 'lng', 'population' ], right_on = [ 'state', 'city', 'event_city', 'lat', 'lng', 'population'])
new_df
new_df = new_df.drop([ 'state', 'city', 'lat', 'lng', 'population'] , axis ="columns")
new_df['region_id'] = new_df['region_id'].astype(int)
new_df
```

```python
test3 = test3.drop(["event_city"], axis = 'columns')
test3
```

```python
import re
```

```python
def strip_character(dataCol):
    r = re.compile(r'[^a-zA-Z !@#$%&*_+-=|\:";<>,./()[\]{}\']')
    return r.sub('', dataCol)

datafile="events.csv"
datafile_df=pd.read_csv(datafile)
pd.set_option("display.max_row", None, "display.max_columns", None)
datafile_df["event_name"] = datafile_df["event_name"].apply(strip_character)
datafile_df["group_name"] = datafile_df["group_name"].apply(strip_character)
datafile_df["venue_event_link"] = datafile_df["venue_event_link"].apply(strip_character)
datafile_df["Google_Map_Link"] = datafile_df["Google_Map_Link"].apply(strip_character)
datafile_df["event_street"] = datafile_df["event_street"].apply(strip_character)
#datafile_df = datafile_df.rename(columns={'Unnamed:':'id'})
#datafile_df = datafile_df.drop(['Unnamed'])
datafile_df.to_csv("events.csv",index=False)
datafile_df.head()
```
